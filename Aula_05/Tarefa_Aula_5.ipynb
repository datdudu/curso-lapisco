# Gerando arquivo que irá ser lido pelos métodos

```python
import pandas as pd

import csv
import random

def generate_random_grade():
    return round(random.uniform(5.0, 10.0), 1)

def create_student_data_csv(file_name):
    headers = ["Nome", "Idade", "Hobby", "Nota"]
    data = []

    nomes = ["Joao", "Maria", "Pedro", "Ana", "Carlos", "Mariana", "Fernando", "Beatriz", "Rafael", "Lucia"]
    idades = [18, 19, 20, 21, 22, 23]
    hobbies = ["Video Game", "Futebol", "Programar", "Colecionar", "Viajar", "Cinema", "Fotografia", "Desenhar", "Musica", "Cozinhar"]


    
    for _ in range(15):
        nome = random.choice(nomes)
        idade = random.choice(idades)
        hobby = random.choice(hobbies)
        

        if random.random() < 0.3:
            nota = "N/A"
        else:
            nota = generate_random_grade()

        data.append([nome, idade, hobby, nota])


    df = pd.DataFrame(data, columns = headers)
   
    df.to_csv(file_name, index=False)
    
    print(f"Arquivo '{file_name}' criado com sucesso.")

# Nome do arquivo que será criado
arquivo_csv = "dados_estudantes.csv"

# Criar o arquivo CSV
create_student_data_csv(arquivo_csv)
```
